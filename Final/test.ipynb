```python
def f1(x):
    return 5*x + 6

def f2(x):
    return 3*x - 2

x_list = [i for i in range(11)]

shares1 = [(x, f1(x)) for x in x_list]
shares2 = [(x, f2(x)) for x in x_list]

print(shares1)
print(shares2)

k = 5
shares3 = [(x, f1(x) + k) for x in x_list]

print(shares3)

# plot the shares

import matplotlib.pyplot as plt

plt.plot(x_list, [share[1] for share in shares1], label='f1')
plt.plot(x_list, [share[1] for share in shares2], label='f2')
plt.plot(x_list, [share[1] for share in shares3], label='f1 + k')

plt.legend()
plt.show()

print("K: ", k)
print("intercepts: ", f1(0), f2(0), shares3[0][1])
```
